```python
type_matches = recent_match['typeMatches']
```

```python
type_matches[0]['seriesMatches'][0]['seriesAdWrapper']
```

```python
int_matches = type_matches[0]['seriesMatches'][0]['seriesAdWrapper']
league_matches = type_matches[1]['seriesMatches'][0]['seriesAdWrapper']
dom_matches = type_matches[2]['seriesMatches'][0]['seriesAdWrapper']

```

```python
data_sets = [
    int_matches,
    dom_matches,
    league_matches
]


all_matches = []

for data in data_sets:
    all_matches.extend(data['matches'])


all_teams = []

for match in all_matches:
    t1 = match['matchInfo']['team1']['teamName']
    t2 = match['matchInfo']['team2']['teamName']

    if t1 not in all_teams:
        all_teams.append(t1)
    if t2 not in all_teams:
        all_teams.append(t2)


wins = {team: 0 for team in all_teams}

for match in all_matches:
    status = match['matchInfo']['status']
    winning_team = status.split(' won')[0]
    wins[winning_team] += 1


for team, win_count in wins.items():
    print(team, win_count)
```
