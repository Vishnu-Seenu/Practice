```python
import requests

venue_ids = [11, 15, 19, 22, 24, 26, 27, 29, 31, 32, 33, 34, 35, 36, 40, 41, 42, 43, 46, 47, 50, 51, 55, 76, 80, 81, 84, 87, 88, 100, 103, 113, 115, 130, 151, 154]

headers = {
	"x-rapidapi-key": "7a79b931f4mshd2ae1a5c14b73dcp10dac1jsnd2e8e16a0f54",
	"x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}
venue_data = {}

for vid in venue_ids:
    try:
        url = f"https://cricbuzz-cricket.p.rapidapi.com/venues/v1/{vid}"
        response = requests.get(url, headers=headers, timeout=5)

        print(f"\n===== Venue ID {vid} =====")

        if response.status_code == 200:
            venue_data[vid] = response.json()
            print(venue_data[vid])
        else:
            print("Failed:", response.status_code)

    except Exception as e:
        print("Error for", vid, e)
    
```

```python
import pandas as pd
import re

rows = []

for vid, v in venue_data.items():
    capacity_raw = str(v.get('capacity', '0'))

    
    match = re.search(r'\d[\d,]*', capacity_raw)
    capacity_num = int(match.group().replace(',', '')) if match else 0

    rows.append({
        'Venue': v.get('ground'),
        'City': v.get('city'),
        'Country': v.get('country'),
        'Capacity': capacity_num
    })

df = pd.DataFrame(rows)


df = df.sort_values('Capacity', ascending=False)

df

```
