```python

import requests


headers = {
	"x-rapidapi-key": "7a79b931f4mshd2ae1a5c14b73dcp10dac1jsnd2e8e16a0f54",
	"x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

allrounders_data = {}

for vid in team_all_ids:
    try:
        url = f"https://cricbuzz-cricket.p.rapidapi.com/stats/v1/player/{vid}/bowling"
        response = requests.get(url, headers=headers, timeout=5)

        print(f"\n===== Venue ID {vid} =====")

        if response.status_code == 200:
            allrounders_data[vid] = response.json()
            print(allrounders_data[vid])
        else:
            print("Failed:", response.status_code)

    except Exception as e:
        print("Error for", vid, e)
```

```python
def extract_runs_wickets(player_data):
    runs = {}
    wickets = {}

    for row in player_data['values']:
        metric = row['values'][0]
        if metric == 'Runs':
            runs = {
                'Test': int(row['values'][1]),
                'ODI': int(row['values'][2]),
                'T20': int(row['values'][3]),
                'IPL': int(row['values'][4])
            }
        if metric == 'Wickets':
            wickets = {
                'Test': int(row['values'][1]),
                'ODI': int(row['values'][2]),
                'T20': int(row['values'][3]),
                'IPL': int(row['values'][4])
            }

    return runs, wickets
```

```python
qualified_players = []

for player_id, player_data in allrounders_data.items():
    player_name = player_data['appIndex']['seoTitle'].split(' Profile')[0]

    runs, wickets = extract_runs_wickets(player_data)

    for fmt in ['Test', 'ODI', 'T20', 'IPL']:
        if runs.get(fmt, 0) > 1000 and wickets.get(fmt, 0) > 50:
            qualified_players.append({
                'player_name': player_name,
                'format': fmt,
                'total_runs': runs[fmt],
                'total_wickets': wickets[fmt]
            })
```

```python
import pandas as pd

df = pd.DataFrame(qualified_players)
print(df)
```
