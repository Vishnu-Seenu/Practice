```python
import requests

url = "https://cricbuzz-cricket.p.rapidapi.com/series/v1/archives/international"

headers = {
	"x-rapidapi-key": "7a79b931f4mshd2ae1a5c14b73dcp10dac1jsnd2e8e16a0f54",
	"x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

response = requests.get(url, headers=headers)

print(response.json())
int_series = response.json()
```

```python
series_map = int_series['seriesMapProto']

int_series_ids = []

keywords = [
    'tour of',
    ' vs ',
    ' v ',
    'world cup',
    'ashes'
]

for year_block in series_map:
    for series in year_block['series']:
        name = series['name'].lower()
        if any(k in name for k in keywords):
            int_series_ids.append(series['id'])

print(int_series_ids)
```

```python
import requests


headers = {
	"x-rapidapi-key": "7a79b931f4mshd2ae1a5c14b73dcp10dac1jsnd2e8e16a0f54",
	"x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

batsman_data = {}

for vid in int_series_ids:
    try:
        url = f"https://cricbuzz-cricket.p.rapidapi.com/series/v1/{vid}"
        response = requests.get(url, headers=headers, timeout=5)

        print(f"\n===== Series ID {vid} =====")

        if response.status_code == 200:
            batsman_data[vid] = response.json()
            print(batsman_data[vid])
        else:
            print("Failed:", response.status_code)

    except Exception as e:
        print("Error for", vid, e)
```

```python
batsman_data
```

```python
venue_ids = set()

for series_data in batsman_data.values():
    for item in series_data.get("matchDetails", []):
        if "matchDetailsMap" not in item:
            continue

        for match in item["matchDetailsMap"].get("match", []):
            venue = match.get("matchInfo", {}).get("venueInfo", {})

            venue_id = venue.get("venueId") or venue.get("id")
            if venue_id:
                venue_ids.add(venue_id)

print("Total unique venue IDs:", len(venue_ids))
print(venue_ids)
```
