```python
all_batsman_ids = ind_batsman_ids + eng_batsman_ids + aus_batsman_ids
```

```python
import requests


headers = {
	"x-rapidapi-key": "7a79b931f4mshd2ae1a5c14b73dcp10dac1jsnd2e8e16a0f54",
	"x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

batsman_data = {}

for vid in all_batsman_ids:
    try:
        url = f"https://cricbuzz-cricket.p.rapidapi.com/stats/v1/player/{vid}/batting"
        response = requests.get(url, headers=headers, timeout=5)

        print(f"\n===== Batsman ID {vid} =====")

        if response.status_code == 200:
            batsman_data[vid] = response.json()
            print(batsman_data[vid])
        else:
            print("Failed:", response.status_code)

    except Exception as e:
        print("Error for", vid, e)
```

```python
def extract_batting_data(player_data):
    runs = {}
    innings = {}

    for row in player_data['values']:
        metric = row['values'][0]

        if metric == 'Runs':
            runs = {
                'Test': int(row['values'][1]),
                'ODI': int(row['values'][2]),
                'T20': int(row['values'][3])
            }

        if metric == 'Innings':
            innings = {
                'Test': int(row['values'][1]),
                'ODI': int(row['values'][2]),
                'T20': int(row['values'][3])
            }

    return runs, innings
```

```python
results = []

for player_id, player_data in batsman_data.items():
    player_name = player_data['appIndex']['seoTitle'].split(' Profile')[0]

    runs, innings = extract_batting_data(player_data)

  
    formats_played = sum(
        1 for fmt in ['Test', 'ODI', 'T20']
        if innings.get(fmt, 0) > 0
    )

    if formats_played < 2:
        continue  

    total_runs = sum(runs.values())
    total_innings = sum(i for i in innings.values() if i > 0)

    overall_avg = round(
        total_runs / total_innings, 2
    ) if total_innings > 0 else 0

    results.append({
        'player_name': player_name,
        'test_runs': runs['Test'],
        'odi_runs': runs['ODI'],
        't20_runs': runs['T20'],
        'overall_batting_avg': overall_avg
    })
```

```python
import pandas as pd

df = pd.DataFrame(results)
df.sort_values(by='overall_batting_avg', ascending=False)
```
